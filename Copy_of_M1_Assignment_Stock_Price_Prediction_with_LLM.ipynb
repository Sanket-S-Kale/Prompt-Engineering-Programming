## Step 3.1: Model Building - Linear Regression for Price Prediction

I will train a regression model to predict tomorrow's stock price using the specified features.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import ta # Ensure 'ta' library is installed for technical indicators
import pandas as pd

# Flatten MultiIndex columns if they exist (common with yfinance output)
if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# Recalculate features to ensure they are present in the 'data' DataFrame
# Calculate Moving Averages
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_10'] = data['Close'].rolling(window=10).mean()
data['MA_50'] = data['Close'].rolling(window=50).mean()

# Calculate RSI
data['RSI'] = ta.momentum.RSIIndicator(data['Close'], window=14).rsi()

# Calculate MACD
macd = ta.trend.MACD(data['Close'])
data['MACD'] = macd.macd()
data['MACD_Signal'] = macd.macd_signal()

# Calculate ATR (Volatility)
data['ATR'] = ta.volatility.AverageTrueRange(data['High'], data['Low'], data['Close'], window=14).average_true_range()

# Add Time-based features
data['Year'] = data.index.year
data['Month'] = data.index.month
data['Day'] = data.index.day
data['DayOfWeek'] = data.index.dayofweek

# Create the target variable for regression: tomorrow's close price
data['Tomorrow_Close_Regression'] = data['Close'].shift(-1)

# Drop rows with NaN values that result from feature calculations and the shifting for the target variable
data_regression = data.dropna().copy()

# Define features (X) and target (y) for regression
feature_cols = ['Close', 'High', 'Low', 'Open', 'Volume', 'MA_5', 'MA_10', 'MA_50', 'RSI', 'MACD', 'MACD_Signal', 'ATR', 'Year', 'Month', 'Day', 'DayOfWeek']

# Filter out features that might not be present if the 'ta' library wasn't installed or data was too short
feature_cols = [col for col in feature_cols if col in data_regression.columns]

X = data_regression[feature_cols]
y = data_regression['Tomorrow_Close_Regression']

# Split data into training and testing sets (70/30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

print(f"Training set size: {len(X_train)} samples")
print(f"Testing set size: {len(X_test)} samples")

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nMean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.3f}")

if r2 < 0:
    print("\nThese metrics indicate that the model is not performing well, as an R² score close to zero or negative suggests that the model does not explain the variability in the target variable better than a simple mean of the target variable. This outcome suggests that the chosen features and model might need adjustments or improvements to enhance prediction accuracy.")
elif r2 < 0.5:
    print("\nThe R-squared value indicates how much of the variance in the dependent variable can be explained by the independent variables. A higher R-squared value indicates a better fit for the model. However, an R-squared below 0.5 might suggest that the model only moderately explains the variance, and further improvements could be beneficial.")
else:
    print("\nThe R-squared value indicates how much of the variance in the dependent variable can be explained by the independent variables. A higher R-squared value indicates a better fit for the model.")
```

## Step 3.2: Model Building - Random Forest Classification

I will now create a binary target variable and train a Random Forest model for classification.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# Ensure MultiIndex columns are flattened if they exist
if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# Create the binary target variable for classification
# 1 if tomorrow's close price is higher than today's, 0 otherwise
data['Tomorrow_Close_Classification'] = data['Close'].shift(-1)
data['target_classification'] = (data['Tomorrow_Close_Classification'] > data['Close']).astype(int)

# Drop rows with NaN values that result from the shifting for the target variable
data_classification = data.dropna().copy()

# Define features (X) and target (y) for classification
feature_cols = ['Close', 'High', 'Low', 'Open', 'Volume', 'MA_5', 'MA_10', 'MA_50', 'RSI', 'MACD', 'MACD_Signal', 'ATR', 'Year', 'Month', 'Day', 'DayOfWeek']

# Filter out features that might not be present
feature_cols = [col for col in feature_cols if col in data_classification.columns]

X_clf = data_classification[feature_cols]
y_clf = data_classification['target_classification']

# Split data into training and testing sets (70/30 ratio), maintaining the chronological order
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.3, shuffle=False)

print(f"Training set size (Classification): {len(X_train_clf)} samples")
print(f"Testing set size (Classification): {len(X_test_clf)} samples")

# Initialize and train the Random Forest Classifier model
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_train_clf, y_train_clf)

# Make predictions on the test set
y_pred_clf = clf_model.predict(X_test_clf)

# Evaluate the model
accuracy = accuracy_score(y_test_clf, y_pred_clf)
conf_matrix = confusion_matrix(y_test_clf, y_pred_clf)
class_report = classification_report(y_test_clf, y_pred_clf)

print(f"\nAccuracy: {accuracy:.2f}")
print(f"\nConfusion Matrix:\n{conf_matrix}")
print(f"\nClassification Report:\n{class_report}")

# Report Feature Importance
feature_importance = pd.DataFrame({'Feature': feature_cols, 'Importance': clf_model.feature_importances_})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

print("\nFeature Importance:")
display(feature_importance)
```

## Step 3.3: Hyperparameter Tuning for Random Forest Classification

I will use `GridSearchCV` to find the optimal hyperparameters for the Random Forest Classifier to improve its performance.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the parameter grid to search
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_features': ['sqrt', 'log2'],
    'max_depth': [5, 10, None],
    'criterion': ['gini', 'entropy']
}

# Initialize a Random Forest Classifier
rf_base = RandomForestClassifier(random_state=42)

# Fit GridSearchCV to the training data
grid_search = GridSearchCV(estimator=rf_base, param_grid=param_grid,
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit GridSearchCV to the training data
grid_search.fit(X_train_clf, y_train_clf)

print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.2f}")

# Get the best model
best_rf_model = grid_search.best_estimator_

# Make predictions on the test set with the best model
y_pred_tuned = best_rf_model.predict(X_test_clf)

# Evaluate the tuned model
accuracy_tuned = accuracy_score(y_test_clf, y_pred_tuned)
conf_matrix_tuned = confusion_matrix(y_test_clf, y_pred_tuned)
class_report_tuned = classification_report(y_test_clf, y_pred_tuned)

print(f"\nOptimized Model Accuracy: {accuracy_tuned:.2f}")
print(f"\nOptimized Model Confusion Matrix:\n{conf_matrix_tuned}")
print(f"\nOptimized Model Classification Report:\n{class_report_tuned}")

# Report Feature Importance from the best model
feature_cols = ['Close', 'High', 'Low', 'Open', 'Volume', 'MA_5', 'MA_10', 'MA_50', 'RSI', 'MACD', 'MACD_Signal', 'ATR', 'Year', 'Month', 'Day', 'DayOfWeek'] # Update feature_cols for display
feature_importance_tuned = pd.DataFrame({'Feature': feature_cols, 'Importance': best_rf_model.feature_importances_})
feature_importance_tuned = feature_importance_tuned.sort_values(by='Importance', ascending=False)

print("\nOptimized Feature Importance:")
display(feature_importance_tuned)
```

## Step 4: Model Evaluation & Prediction (Optimized Random Forest)

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# Make predictions on the test set with the best tuned model
y_pred_tuned_final = best_rf_model.predict(X_test_clf)

# Evaluate the tuned model
accuracy_tuned_final = accuracy_score(y_test_clf, y_pred_tuned_final)
conf_matrix_tuned_final = confusion_matrix(y_test_clf, y_pred_tuned_final)
class_report_tuned_final = classification_report(y_test_clf, y_pred_tuned_final)

print(f"\nOptimized Model Accuracy for Prediction Period: {accuracy_tuned_final:.2f}")
print(f"\nOptimized Model Confusion Matrix for Prediction Period:\n{conf_matrix_tuned_final}")
print(f"\nOptimized Model Classification Report for Prediction Period:\n{class_report_tuned_final}")
```
